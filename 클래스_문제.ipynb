#### 도서 관리 시스템 구현하기.  
Book 클래스:
- title: 책 제목
- author: 저자
- isbn: ISBN 번호
-  available: 책의 대여 가능 여부 (True/False) , 기본적으로 책은 대여 가능
- \_\_str\_\_() 메소드: 책의 정보를 문자열로 반환

Library 클래스:
- books: Book 객체를 저장하는 리스트
- add_book(book): 책을 추가하는 메소드
- remove_book(isbn): ISBN으로 책을 제거하는 메소드
- find_book(isbn): ISBN으로 책을 찾는 메소드
- list_books(): 모든 책의 정보를 출력하는 메소드

제약 조건:
- ISBN 번호는 고유해야 하며, 중복된 ISBN으로 책을 추가할 수 없습니다.

책추가
- book1 = Book("파이썬 프로그래밍", "홍길동", "1234")
- book2 = Book("자바 프로그래밍", "김철수", "0987")


```python
class Book :
  def __init__(self,title,author,isbn) :
    self.title = title
    self.author = author
    self.isbn = isbn
    self.available = True
  def __str__(self) :
    return f'제목:{self.title},저자:{self.author},ISBN:{self.isbn},대여가능:{self.available}'
```

```python
book1 = Book("파이썬 프로그래밍", "홍길동", "1")
book2 = Book("자바 프로그래밍", "김철수", "2")
```

```python
print(book1,book2)
```

```python
class Library :

  def __init__(self) :
    self.books = []

  # 도서 추가
  def add_book(self,book) :
    for ele in self.books :
      if ele.isbn == book.isbn :
        print('isbn 중복!')
        return

    self.books.append(book)
    print(f'{book.title}가 추가됨!')

  #도서 제거
  def remove_book(self,isbn) :
    for ele in self.books :
      if ele.isbn == isbn :
        self.books.remove(ele)
        print(f'{ele.title}가 제거됨!')
        return
    print(f'ISBN = {isbn} 도서를 찾을수 없음!')

  #도서 검색
  def find_book(self,isbn) :
    for ele in self.books :
      if ele.isbn == isbn :
        return ele
    print(f'ISBN = {isbn} 도서를 찾을수 없음!')

  #도서 목록
  def list_books(self) :
    for ele in self.books :
      print(ele)
```

```python
#도서 생성
book1 = Book("파이썬 프로그래밍", "홍길동", "1")
book2 = Book("자바 프로그래밍", "김철수", "2")
```

```python
#도서관 생성
lib = Library()
```

```python
lib.add_book(book1)
lib.add_book(book2)
```

```python
lib.list_books()
```

#### 예외처리 반영하기
- 도서관리 시스템에서 중복저장과, 도서검색이 안된경우에 대해 예외처리 적용해보기

```python
class Book :
  def __init__(self,title,author,isbn) :
    self.title = title
    self.author = author
    self.isbn = isbn
    self.available = True
  def __str__(self) :
    return f'제목:{self.title},저자:{self.author},ISBN:{self.isbn},대여가능:{self.available}'
```

```python
class Library :

  def __init__(self) :
    self.books = []

  # 도서 추가
  def add_book(self,book) :
    for ele in self.books :
      if ele.isbn == book.isbn :
        # print('isbn 중복!')
        raise ValueError('isbn 중복!')
        return

    self.books.append(book)
    print(f'{book.title}가 추가됨!')

  #도서 제거
  def remove_book(self,isbn) :
    for ele in self.books :
      if ele.isbn == isbn :
        self.books.remove(ele)
        print(f'{ele.title}가 제거됨!')
        return
    # print(f'ISBN = {isbn} 도서를 찾을수 없음!')
    raise ValueError(f'ISBN = {isbn} 도서를 찾을수 없음!')

  #도서 검색
  def find_book(self,isbn) :
    for ele in self.books :
      if ele.isbn == isbn :
        return ele
    # print(f'ISBN = {isbn} 도서를 찾을수 없음!')
    raise ValueError(f'ISBN = {isbn} 도서를 찾을수 없음!')

  #도서 목록
  def list_books(self) :
    for ele in self.books :
      print(ele)
```

```python
#도서 생성
book1 = Book("파이썬 프로그래밍", "홍길동", "1")
book2 = Book("자바 프로그래밍", "김철수", "2")
```

```python
#도서 목록
library.list_books()
```
