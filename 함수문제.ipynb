#### 문제 1: 매개변수의 타입 검사
- 문제: 두 개의 숫자를 받아서 그 합계를 반환하는 함수를 작성하되,
입력된 매개변수가 숫자인지 확인하고, 숫자가 아닐 경우 ValueError를 발생시키도록 하세요.


```python
def add_numbers(a, b):
    # if not isinstance(a, (int, float)) or not isinstance(b, (int, float)):
    if not isinstance(a, int|float) or not isinstance(b, int|float):     # union type
        raise ValueError("Both a and b must be numbers.")
    return a + b

# 테스트 예시
print(add_numbers(5, 3))  # 8
# print(add_numbers(5, '3')) # ValueError
```

#### 문제 2: 기본값과 가변 매개변수 혼합
- 문제: 기본값을 가지는 매개변수와 가변 매개변수를 사용하여, 특정 문자열을 여러 번 반복하는 함수를 작성하세요. 기본값은 1로 설정하고, 가변 매개변수로 추가 문자열을 받을 수 있어야 합니다.

```python
def repeat_string(string, n=1, *args):
    result = string * n
    for arg in args:
        result += arg
    return result
```

```python
# 테스트 예시
print(repeat_string('kh', 3))
print(repeat_string('kh',3, '인재교육원','KDT과정'))
print(repeat_string('kh'))
```

#### 문제 4: 매개변수의 언패킹
- 문제: 딕셔너리를 받아서 키와 값을 출력하는 함수를 작성하세요. 이 함수는 키워드 매개변수를 사용하여 딕셔너리의 내용을 언패킹해야 합니다.

```python
def print_dict(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")
```

```python
# 테스트
print_dict(a=1, b=2, c=3)
print_dict(name='홍길동', age=30)
```

```python
person = {'name': '홍길동', 'age': 30}
print_dict(**person)
```

```python
print_dict(**{'name': '홍길동', 'age': 30})
```

문제 5: 위치 매개변수와 키워드 전용 매개변수 혼합 사용하기
- 문제: 두 개의 숫자와 연산자(문자열)를 받아서 계산 결과를 반환하는 함수를 작성하되, 연산자는 키워드 전용 매개변수로 설정하라.(4칙연산)
- 다중IF문
- match문

```python
def compute(a, b, *, operator):
    if operator == '+':
        return a + b
    elif operator == '-':
        return a - b
    elif operator == '*':
        return a * b
    elif operator == '/':
      if b == 0:
        raise ValueError("피젯수가 0 입니다!")
      return a / b
    else:
       raise ValueError("지원하지 않는 연산!")
```

```python
# 테스트
print(compute(5, 3, operator='+'))
print(compute(5, 3, operator='*'))
```

```python
print(compute(5, 0, operator='/'))
```

```python
def compute2(a, b, *, operator):
  match operator:
    case '+':
      return a + b
    case '-':
      return a - b
    case '*':
      return a * b
    case '/':
      if b == 0:
        raise ValueError("피젯수가 0 입니다!")
      return a / b
    case '-':
      raise ValueError("지원하지 않는 연산!")
```

```python
# 테스트
print(compute(5, 3, operator='+'))
print(compute(5, 3, operator='*'))
```

```python
print(compute(5, 0, operator='/'))
```

```python
print(compute(5, 0, operator='//'))
```
