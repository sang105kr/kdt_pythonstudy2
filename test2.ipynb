```python
def describe_shape(shape):
  match shape:
    case 'triangle':
      print('삼각형입니다.')
    case 'square':
      print('정사각형입니다.')
    case 'circle':
      print('원입니다.')
    case _ :
      return "Unknown shape"
describe_shape('circle')
```

```python
def describe_shape(shape):
  match shape:
    case ('triangle',base,height):
      print('삼각형입니다.')
    case ('square',width,height):
      print('정사각형입니다.')
    case ('circle',radius):
      print('원입니다.')
    case _ :
      return "Unknown shape"
```

```python
describe_shape('aaa')
```

```python
describe_shape(('triangle',10,20))
```

### 리스트 컴프리헨션
- 리스트나 다른 반목 가능한 객체(iterable)에서 조건을 적용하거나 변형하여 새로운 리스트를 생성하는 방법

```python
numbers = [1,2,3,4,5]
# type(numbers)
new_list = ["짝수" if x % 2 == 0 else "홀수" for x in numbers ]
print(new_list)
```
