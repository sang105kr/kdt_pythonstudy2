- items() : 딕서너리의 키-값 쌍을 반환
- keys() : 딕셔너리의 키를 반환
- values() : 딕셔너리의 값을 반환

```python
for k,v in d.items() :
  print(f'key={k},value={v}')
```

```python
for k in d.keys() :
  print(f'key={k},value={d[k]}')
```

```python
for v in d.values() :
  print(f'value={v}')
```

```python
print(type(d.items()))
print(type(d.keys()))
print(type(d.values()))
```

키로 값져오기
- get(key, default=None) : key로 value를 반환, 키값이 없어도 오류가 발생하지 않는다. 키값이 없을때 기본값을 지정할 수 있다.
- dict[key] : 키값이 없으면 오류 발생

```python
d.get('name'), d['name']
```

```python
d.get('hoby')
```

```python
d['hoby']
```

```python
d.get('hoby','등산')
```

아이템 추가
- update(other_dict) : 동일한키가 없으면 추가, 동일한키가 있으면 업데이트
- dict[key] = value :동일한키가 없으면 추가, 동일한키가 있으면 업데이트
- setdefulat(key, default=None) : 키가 존재하지 않으면 default값을 추가하고 값을 반환, 키가 존재하면 키의 값을 반환

```python
d = {'name':'홍길동','age':20,'gender':'man'}
d.update({'name':'홍길순','region':'울산','tel':'010-1234-5678'})
d
```

```python
d = {'name':'홍길동','age':20,'gender':'man'}
d['region'] = '울산'
d
```

```python
d = {'name':'홍길동','age':20,'gender':'man'}
d.setdefault('name','홍길순') # 키가 존재하면 해당 키의 값을 반환
```

```python
d = {'name':'홍길동','age':20,'gender':'man'}
d.setdefault('region','울산') # 키가 존재하지 않으면 키와 default값을 쌍으로하여 추가
d
```

아이템 제거

- pop(key, defalut=None) : 키에대 한 값을 반환하고 해당 아이템을 삭제, 키가 없으면 default값을 반환, default값이 없으면 KeyError가 발생
- popitem() : 딕셔너리의 마지막 아이템을 반환하고 삭제, 딕셔너리가 비어있으면 KeyError


```python
d = {'name':'홍길동','age':20,'gender':'man'}
removedItem = d.pop('gender')
print(removedItem)
print(d)
```

```python
try :
  removedItem = d.pop('gender2')
except KeyError as err :
  print(err)
  print(dir(err))
```

```python
d = {'name':'홍길동','age':20,'gender':'man'}
removed_item = d.popitem()
print(removed_item)
print(d)
```
