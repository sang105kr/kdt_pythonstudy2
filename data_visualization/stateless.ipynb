#### Stateless 방식
1. 사용자가 Figure Axes 객체를 직접 생성
2. oop방식
3. 복잡한 레이아웃을 시각화하거나 세밀한 프로팅을 할때

```python
fig, ax = plt.subplots(figsize=(6,4))
ax.plot([1,2,3,4],[1,4,2,3])
ax.set_title('제목')
ax.set_xlabel('X축라벨')
ax.set_ylabel('y축라벨')
plt.show()
```

```python
fig, axes = plt.subplots(2,2,figsize=(6,4))
axes[1][0].plot([1,2,3,4],[1,4,2,3])
axes[1][0].set_title('제목')
axes[1][0].set_xlabel('X축라벨')
axes[1][0].set_ylabel('y축라벨')

axes[0][1].scatter([1,2,3,4],[1,4,2,3])
axes[0][1].set_title('제목')
axes[0][1].set_xlabel('X축라벨')
axes[0][1].set_ylabel('y축라벨')

axes[0][0].bar([1,2,3,4],[1,4,2,3])
axes[0][0].set_title('제목')
axes[0][0].set_xlabel('X축라벨')
axes[0][0].set_ylabel('y축라벨')

data = pd.Series([1,1,2,2,2,3,4,6])
axes[1][1].hist(data, bins=3)
axes[1][1].set_title('제목')
axes[1][1].set_xlabel('X축라벨')
axes[1][1].set_ylabel('y축라벨')
plt.show()
```

### 데이터 프레임에서 플로팅

```python
import random

random.seed(1)

data = { 
  '국어' : random.sample(range(60,100+1),5),
  '수학' : random.sample(range(60,100+1),5),
  '영어' : random.sample(range(60,100+1),5) 
}
data
```

```python
df = pd.DataFrame(data=data, index=['홍1','홍2','홍3','홍4','홍5'])
df
```

```python
df.plot()
```

```python
fig, axes = plt.subplots(2,2)
fig.suptitle('국/영/수 점수 시각화') # Figure의 제목

df['국어'].plot(ax=axes[0][0],kind='line')
df['수학'].plot(ax=axes[1][0],kind='bar')
df['영어'].plot(ax=axes[1][1],kind='hist', bins=3)
df.plot(x='영어',y='수학',ax=axes[0][1],kind='scatter')

axes[0][0].set_title('국어')
axes[0][0].set_xlabel('이름')
axes[0][0].set_ylabel('점수')

axes[1][0].set_title('수학')
axes[0][0].set_xlabel('이름')
axes[0][0].set_ylabel('점수')

axes[1][1].set_title('영어')
axes[0][0].set_xlabel('점수')
axes[0][0].set_ylabel('빈도')

axes[0][1].set_title('영어/수학 상관관계')

plt.tight_layout()
plt.show()
```
