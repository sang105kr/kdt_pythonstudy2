튜플 : 리스트의 불변 객체

```python
numbers = (1,2,3,4,5)
numbers.append(6)
numbers
```

```python
numbers = (1,2,3,4,5)
numbers.remove(3)
numbers
```

```python
numbers = (1,2,3,4,5)
numbers.clear(3)
numbers
```

```python
numbers = (1,2,3,4,5)
numbers.insert(3,10)
numbers
```

```python
numbers[2] = 10
```

```python
numbers.count(2)
```

```python
numbers = (1,2,3,4,5)
copied_numbers = numbers.copy()
print(copied_numbers,numbers)
```

```python
numbers = (1,2,3,4,5)
idx = numbers.index(3)
print(idx)
print(max(numbers),min(numbers))
print(len(numbers))
```
