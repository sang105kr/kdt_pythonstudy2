```python
from abc import abstractmethod
# 추상클래스
class Person() :
  def __init__(self) :
    raise Exception('추상클래스!')

  @abstractmethod
  def sleep(self):
    pass

  def smile(self) :
    print('웃다')

  def eat(self) :
    print('먹다')
```

```python
# Person 타입의 객체생성 불가
p = Person('홍길동',20)
id(p)
```

```python
class Student(Person) :
  def __init__(self,name,age,major) :
    self.name = name
    self.age = age
    self.major = major

  def sleep(self) :
    print(f'{self.name} 가 자다')

  def eat(self) :
    print(f'{self.name} 가 먹다')
```

```python
s1 = Student('홍길동',10,'컴공')
print(id(s1))
s1.smile()
s1.eat()
s1.sleep()
```

```python
class Person :
  def __init__(self,name,age) :
    self.name = name
    self.age = age

  def get_name(self):
    return self.name

  def set_name(self,name):
    self.name = name
```

```python
p1 = Person('홍길동',20)
p1.get_name()
p1.set_name('홍길순')
p1.get_name()
```

#### getter, setter 구현
- 파이썬에서는 getter,setter를 메소드 호출방식이 아니라 속성 접근하듯이 사용가능

```python
class Person :
  def __init__(self,name,age) :
    self._name = name
    self._age = age

  @property
  def name(self):
    return self._name + '님'

  @name.setter
  def name(self,name):
    self._name = name
```

```python
p1 = Person('홍길동',20)
print(p1.name)
p1.name = '홍길순'
print(p1.name)
```
