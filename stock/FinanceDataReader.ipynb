```python
# 데이터 프레레임의 컬럼 메타정보
stocks.columns
```

```python
stocks_columns = ['Code', 'Name', 'Market', 'Dept','Close', 'ChangeCode','Changes', 'ChagesRatio', 'Open', 'High', 'Low', 'Volume', 'Amount',
       'Marcap', 'Stocks', 'MarketId'] 
```

```python
stocks_desc_columns = ['Code', 'Sector', 'Industry', 'ListingDate','SettleMonth', 'Representative', 'HomePage', 'Region']
```

```python
stocks = stocks[stocks_columns]
```

```python
stocks_desc = stocks_desc[stocks_desc_columns]
```

```python
import numpy as np
import pandas as pd
```

```python
#stocks join stocks_desc
new_stocks = pd.merge(stocks,stocks_desc,on='Code')
```
